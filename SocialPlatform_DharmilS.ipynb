# Altair is a declarative statistical visualization library for Python, based on Vega and Vega-Lite. Altair offers a powerful and concise visualization grammar that enables you to build a wide range of statistical visualizations quickly. 

This is a data set that contains number of active users (updated:29th April 2019) of Twitter, Facebook, LinkedIn, Snapchat, and Instagram of countries, United States, United Kingdom, India, Brazil, Mexico. Data was pulled from DataReportal and cleaned by Dharmil Sanghavi. All the values of active users are in millions. Here I am demonstrating how to use Altair to create some basic bar charts.

```python
import altair as alt
alt.renderers.enable('notebook')
```

```python
#Source:https://datareportal.com/reports/digital-2019-q2-global-digital-statshot, Data Cleaned by Dharmil Sanghavi, value is in millions
data = alt.Data(values = 
           [{ 'Country': 'INDIA', 'Platform': 'FACEBOOK', 'value': '260' },
                          { 'Country': 'INDIA', 'Platform': 'INSTAGRAM', 'value': '64' },
                          { 'Country': 'INDIA', 'Platform': 'LINKEDIN', 'value': '56' },
                          { 'Country': 'INDIA', 'Platform': 'SNAPCHAT', 'value': '12.15' },
                          { 'Country': 'INDIA', 'Platform': 'TWITTER', 'value': '7.85' },
                          { 'Country': 'UNITED KINGDOM', 'Platform': 'FACEBOOK', 'value': '37' },
                          { 'Country': 'UNITED KINGDOM', 'Platform': 'INSTAGRAM', 'value': '22' },
                          { 'Country': 'UNITED KINGDOM', 'Platform': 'LINKEDIN', 'value': '26' },
                          { 'Country': 'UNITED KINGDOM', 'Platform': 'SNAPCHAT', 'value': '16.95' },
                          { 'Country': 'UNITED KINGDOM', 'Platform': 'TWITTER', 'value': '14.10' },
                          { 'Country': 'UNITED STATES', 'Platform': 'FACEBOOK', 'value': '190' },
                          { 'Country': 'UNITED STATES', 'Platform': 'INSTAGRAM', 'value': '110' },
                          { 'Country': 'UNITED STATES', 'Platform': 'LINKEDIN', 'value': '150' },
                          { 'Country': 'UNITED STATES', 'Platform': 'SNAPCHAT', 'value': '93.55' },
                          { 'Country': 'UNITED STATES', 'Platform': 'TWITTER', 'value': '49.45' },
                          { 'Country': 'BRAZIL', 'Platform': 'FACEBOOK', 'value': '120' },
                          { 'Country': 'BRAZIL', 'Platform': 'INSTAGRAM', 'value': '66' },
                          { 'Country': 'BRAZIL', 'Platform': 'LINKEDIN', 'value': '36' },
                          { 'Country': 'BRAZIL', 'Platform': 'SNAPCHAT', 'value': '9.80' },
                          { 'Country': 'BRAZIL', 'Platform': 'TWITTER', 'value': '8.57' },
                          { 'Country': 'MEXICO', 'Platform': 'FACEBOOK', 'value': '79' },
                          { 'Country': 'MEXICO', 'Platform': 'INSTAGRAM', 'value': '20' },
                          { 'Country': 'MEXICO', 'Platform': 'LINKEDIN', 'value': '13' },
                          { 'Country': 'MEXICO', 'Platform': 'SNAPCHAT', 'value': '10.95' },
                          { 'Country': 'MEXICO', 'Platform': 'TWITTER', 'value': '7.09' }
                         ])
```

```python
oi_scale = alt.Scale(domain=['INDIA', 'UNITED KINGDOM', 'UNITED STATES', 'BRAZIL', 'MEXICO'],
                  range=['#0072B2', '#D55E00', '#009E73', '#CC79A7', '#E69F00'])
```

```python
#Horizontal Stacked bar chart 
alt.Chart(data).mark_bar().encode(
    alt.X('value:Q'),
    alt.Y('Platform:N'),
    alt.Color('Country:N',scale=oi_scale))
```

```python
#Verical Stacked Bar Chart
alt.Chart(data).mark_bar().encode(
    alt.X('Platform:N'),
    alt.Y('value:Q'),
    alt.Color('Country:N',scale=oi_scale))
```

```python
#Horizontal Layered Bar Chart
alt.Chart(data).mark_bar(opacity=0.7).encode(
    alt.X('value:Q'),
    alt.Y('Platform:N'),
    alt.Color('Country:N',scale=oi_scale),
)
```

```python
#Vertical Layered Bar Chart 
alt.Chart(data).mark_bar(opacity=0.7).encode(
    alt.X('Platform:N'),
    alt.Y('value:Q'),
    alt.Color('Country:N',scale=oi_scale),
)
```

```python
# create grouped bar chart (By Platform)
alt.Chart(data).mark_bar(size=20).encode(
    alt.Y('value:Q'),
    alt.X('Country:N'),
    alt.Color('Country:N',scale=oi_scale),
    alt.Column('Platform:N') # facet by Platform
).properties(
    width=130,
    height=150
)
```

```python
# create grouped bar chart (By Country)
alt.Chart(data).mark_bar(size=20).encode(
    alt.X('value:Q'),
    alt.Y('Platform:N'),
    alt.Color('Country:N',scale=oi_scale),
    alt.Column('Country:N') # facet by Country
).properties(
    width=130,
    height=150
)
```

```python
#Normalized Bar Chart
alt.Chart(data).mark_bar(opacity=0.7).encode(
    alt.X('value:Q', stack="normalize"),
    alt.Y('Platform:N'),
    alt.Color('Country:N',scale=oi_scale),
)
```

```python
#Interactive Scatterplot 
alt.Chart(data).mark_circle().encode(
    alt.X('value:Q'),
    alt.Y('Platform:N'),
    alt.Color('Country:N',scale=oi_scale),
).interactive()
```
